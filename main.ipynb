<h1 style="font-weight: bold;">Mengubah data menjadi DataFrame<h1>

```python
import re
import pandas as pd
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import numpy as np
import math

df = pd.read_excel("Dataset B.xlsx")
pd.set_option('display.max_colwidth', None)
data_ori = df.iloc[110:120]
data = data_ori
data
```

<h1 style="font-weight: bold;">Menghilangkan angka, emoji, kapital, spasi, whitespace, dan enter<h1>

```python
def case_folding(sentence):
    sentence = re.sub(r'[0-9]', ' ', sentence)
    sentence = re.sub(r'[^\w\s]', ' ', sentence)
    sentence = sentence.lower()
    sentence = sentence.strip()
    sentence = re.sub(r'\s+', ' ', sentence)
    sentence = sentence.replace('\n', '')
    return sentence

data = data['Reviews'].tolist()
data = [case_folding(sentence) for sentence in data]
print(*data, sep='\n')
```

<h1 style="font-weight: bold;">Memisahkan setiap kata dalam kalimat<h1>

```python
def tokenizing(sentence):
    return sentence.split()

data = [tokenizing(sentence) for sentence in data]
print(*data, sep='\n')
```

<h1 style="font-weight: bold;">Menghilangkan stopwords<h1>

```python
my_stopwords_dictionary = [
    'nya',
    'masa',
    'doang',
    'banget',
    'bangettt',
    'harusnya',
    'yg',
    'lama',
    'datang',
    'cuma',
    'gak',
    'mau',
    'tu',
    'bos',
    'aja',
    'kok',
    'tau',
    'd',
    'ap',
    'smpai',
    'apa',
    'si',
    'sy',
    'sja',
    'sih',
    'bnyak',
    'sm',
]

def stopwords(sentence):
    factory = StopWordRemoverFactory()
    stopword = factory.create_stop_word_remover()
    return [stopword.remove(word) for word in sentence if stopword.remove(word) != '' and word not in my_stopwords_dictionary]

data = [stopwords(sentence) for sentence in data]
print(*data, sep='\n')
```

<h1 style="font-weight: bold;">Mengubah menjadi kata baku<h1>

```python
my_root_words = {
    'pengemasan': 'kemas',
    'pengemasannya': 'kemas',
    'kecepetan': 'kecepatan',
    'gede': 'besar',
    'lumayan': 'cukup',
    'dipake': 'dipakai',
    'komlen': 'komplain',
    'julan': 'jualan',
    'nyesal': 'menyesal',
    'pcs': 'picis',
    'kecewaaaa': 'kecewa',
    'datengnya': 'datangnya',
    'dkirim': 'dikirim',
}

def my_stemming(sentence):
    for i in range(len(sentence)):
        if sentence[i] in my_root_words:
            sentence[i] = my_root_words[sentence[i]]
    return sentence

data = [my_stemming(sentences) for sentences in data]
```

```python
def stemming(sentence):
    factory = StemmerFactory()
    stemmer = factory.create_stemmer()
    return [stemmer.stem(word) for word in sentence]

data = [stemming(sentences) for sentences in data]
print(*data, sep='\n')
```

<h1 style="font-weight: bold;">Menghitung jumlah dokumen dan jumlah kata<h1>

```python
def count_doc():
    list = []
    for i in range(len(data)):
        list.append('Doc' + str(i + 1))
    return list

def count_words(sentence):
    list = []
    for sentence in data:
        for word in sentence:
            if word not in list:
                list.append(word)
    return sorted(list)

all_doc = count_doc()
print('Documents:', len(all_doc))

all_words = count_words(data)
print('Words:', len(all_words))
```

<h1 style="font-weight: bold;">One-Hot Encoding<h1>

```python
zeros_matrix_1 = np.zeros((len(all_doc), len(all_words)), dtype=object)

def one_hot_encoding(matrix):
    for i in range(len(all_doc)):
        for j in range(len(all_words)):
            for k in range(len(data[i])):
                if all_words[j] == data[i][k]:
                    matrix[i][j] = 1
    return matrix

df = pd.DataFrame(one_hot_encoding(zeros_matrix_1), columns=all_words, index=all_doc)
df.T
```

<h1 style="font-weight: bold;">Bag-of-words<h1>

```python
zeros_matrix_2 = np.zeros((len(all_doc), len(all_words)), dtype=object)

def one_hot_encoding(matrix):
    for i in range(len(all_doc)):
        for j in range(len(all_words)):
            for k in range(len(data[i])):
                if all_words[j] == data[i][k]:
                    matrix[i][j] += 1
    return matrix

df = pd.DataFrame(one_hot_encoding(zeros_matrix_2), columns=all_words, index=all_doc)
df.T
```

<h1 style="font-weight: bold;">TFIDF<h1>

```python
def tfidf(data):
    tfidf = []
    for i in range(len(data)):
        tf = data[i] / sum(data)
        idf = math.log10(len(data) / (1 + data[i]))
        tfidf.append(tf * idf)
    return tfidf

def tfidf_matrix(matrix):
    tfidf_matrix = []
    for i in range(len(matrix)):
        tfidf_matrix.append(tfidf(matrix[i]))
    return tfidf_matrix

df = pd.DataFrame(tfidf_matrix(zeros_matrix_2), columns=all_words, index=all_doc)
df.T
```
